```python
import pygal
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from IPython.display import display, SVG

# Dataset value student in big data course
data = {
    'Students': ['Andi', 'Budi', 'Citra', 'Dewi', 'Eka', 'Fajar', 'Gita', 'Hadi', 'Indah', 'Joko'],
    'Nilai': [85, 78, 92, 88, 76, 95, 89, 74, 90, 82]
}
df = pd.DataFrame(data)

# 1. Bar Chart using Pygal
bar_chart = pygal.Bar()
bar_chart.title = 'Student Score for Big Data Course'
for index, row in df.iterrows():
    bar_chart.add(row['Students'], row['Nilai'])
bar_chart.render_to_file('bar_chart.svg')
display(SVG('bar_chart.svg'))

# 2. Scatter Plot using Seaborn
plt.figure(figsize=(6, 4))
sns.scatterplot(x=df['Students'], y=df['Nilai'], s=100, color='blue')
plt.title('Student Score Distribution')
plt.xlabel('Students')
plt.ylabel('Score')
plt.xticks(rotation=45)
plt.savefig('scatter_plot.png')
plt.show()

# 3. Line Chart using Matplotlib
plt.figure(figsize=(6, 4))
plt.plot(df['Students'], df['Nilai'], marker='o', linestyle='-', color='red')
plt.title('Progress')
plt.xlabel('Students')
plt.ylabel('Score')
plt.xticks(rotation=45)
plt.grid()
plt.savefig('line_chart.png')
plt.show()
```
